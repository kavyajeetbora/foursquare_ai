# Setup Environment

1. Install Pmtiles
2. GDAL
3. tilemaker

```python
%%shell

# Download go-pmtiles for Linux x86_64 quietly
wget -q https://github.com/protomaps/go-pmtiles/releases/download/v1.28.3/go-pmtiles_1.28.3_Linux_x86_64.tar.gz

# Verify SHA256 checksum
echo "06cf492adc2c7fcd23c4f11a98a5292f4cbe04d3afc3a6b38a07bb47452daca2 go-pmtiles_1.28.3_Linux_x86_64.tar.gz" | sha256sum --check --quiet

# Extract quietly
tar -xzf go-pmtiles_1.28.3_Linux_x86_64.tar.gz -C /tmp/ > /dev/null 2>&1

# Install binary to /usr/local/bin (assuming binary is named 'pmtiles')
sudo mv /tmp/pmtiles /usr/local/bin/pmtiles > /dev/null 2>&1

# Clean up
rm go-pmtiles_1.28.3_Linux_x86_64.tar.gz

# Check installation
echo "pmtiles Installation complete."
pmtiles version

## GDAL Installation
apt-get update -qq
apt-get install -y libgdal-dev gdal-bin python3-gdal
apt-get install -y build-essential
export CPLUS_INCLUDE_PATH=/usr/include/gdal
export C_INCLUDE_PATH=/usr/include/gdal
pip install GDAL==$(gdal-config --version)
echo "GDAL Installation complete."
gdalinfo --version
```

## Installing `tilemaker`

```python
%%shell
sudo apt install build-essential libboost-dev libboost-filesystem-dev libboost-program-options-dev libboost-system-dev lua5.1 liblua5.1-0-dev libshp-dev libsqlite3-dev rapidjson-dev

cd tilemaker
make
sudo make install
```

## Download OSM data for India

```python
%%shell
cd /content
wget -O india.osm.pbf https://download.geofabrik.de/asia/india-latest.osm.pbf
```

```python
%%shell
tilemaker /content/india.osm.pbf --output /content/india-latest.mbtiles --config /content/tilemaker/resources/config-openmaptiles.json --process /content/tilemaker/resources/process-openmaptiles.lua
```
